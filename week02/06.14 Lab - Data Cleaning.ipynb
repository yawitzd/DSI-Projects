```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

bb = pd.read_csv('/Users/TuckerCS/Dropbox/DSI-Projects/week02/billboard.csv')
```

```python
bb.head()
```

```python
## Fixing Names ##

def nameswitch(value):
   if "," in value:
       for i in range(len(value)):
           if value[i] == ",":
               return value[i+2:len(value)] + " " + value[:i] 
   else:
       return value

bb['artist.inverted'] = bb['artist.inverted'].to_frame().applymap(nameswitch)
```

```python
## Fixing Dates ##

bb['date.entered'] = pd.to_datetime(bb['date.entered'])
bb['date.peaked'] = pd.to_datetime(bb['date.peaked'])
```

```python
## Fixing Times ##

def timeswitch(value):
    return '0' + value[0] + ':' + value[2:4]
           
bb['time'] = bb['time'].to_frame().applymap(timeswitch)       
```

```python
## Replacing * with np.nan values ##

def replace_null(value):
    if value == '*':
        return np.nan
    else:
        return value
    
bb = bb.applymap(replace_null)        
```

```python
def countnan(value):
    for i in value:
        if i != (np.nan):
            return pd.value_counts(i)

bb.iloc[:,7:].apply(countnan, axis=1)
```
