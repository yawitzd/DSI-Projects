```python
import pandas as pd
```

```python
def fizzbuzz(n):
    if n%15 == 0:
        return 'FizzBuzz'
    elif n%3 == 0:
        return 'Fizz'
    elif n%5 == 0:
        return 'Buzz'
    else:
        return n
```

```python
nums = [range(0,101)]

df = pd.DataFrame(nums)

df = df.T
```

```python
df['FizzBuzz'] = df.applymap(fizzbuzz)
```
