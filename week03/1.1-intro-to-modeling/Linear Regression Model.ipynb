```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import string
import seaborn as sns
import matplotlib.pyplot as plt
import pylab as pl
%matplotlib inline
plt.style.use('ggplot')
```

```python
values = [(-1,0),(1,3),(2,4),(3,7),(4,10)]

def linear_regression(value):
    x = []
    y = []
    x2 = []
    xy = []
    for i,j in value:
        x.append(i)
        y.append(j)
        x2.append(i**2)
        xy.append(i*j)
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    x2_mean = np.mean(x2)
    xy_mean = np.mean(xy)
    slope = (xy_mean - (x_mean*y_mean)) / (x2_mean - (x_mean**2))
    intercept = y_mean - slope * x_mean
    
    fig, ax = plt.subplots(figsize=(10,8))
    ax.scatter(x,y)
    
    new_y_values = []
    for m in x:
        new_y_values.append(m*slope+intercept)
    ax.plot(x, new_y_values, linewidth=2)
    return slope, intercept
        
linear_regression(values)
```
